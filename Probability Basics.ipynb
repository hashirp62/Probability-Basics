3. A bag has 3 red and 2 blue balls. If one ball is drawn randomly   and is red, what is the probability

 > Number of ball= 3(red) + 2(blue)= 5

   remaining red ball left after first draw

   = 5-1 = 4

   = 2/4

   1/2 = 0.5

   **probability of getting red ball in second draw(without replacement) is 1/2 or 0.5 that is 50%**

5. The average height of 1000 students = 160 cm. A sample of 100 students shows an average height
= 158 cm. Find the sampling error.

 > Sampling error is the difference between the result obtained from a sample and the actual value of the population.

It occurs because only a small part of the population is studied instead of the whole population.

  Here,

  population mean= 160

  sample mean= 158

  sampling error= sample mean - population mean

  = 158-160

  = -2

  **sampling error= -2**

  ** here sampling error is equal to -2. which means sample average height is 2cm less than population height**

6. The population mean salary is ₹50,000 with σ = ₹5,000. If we take a sample of 100 employees, what
is the standard error of the mean (SEM)?

 > Population standard deviation(σ)= 5000

   sample size (n)= 100

   SEM= Population standard deviation(σ)/√sample size (n)

   = 5000/√100

   = 5000/10

   = 500

   **standard error of the mean= ₹500**
